```python
df_org = pd.read_csv('../data/cleaned_ameshousing.csv')
```

```python
df_org.head()
```

```python
columns = ['Overall Qual', 'Year Built', 'Total Bsmt SF', 'Total Full Bath', 'Total Half Bath', 'Gr Liv Area', 'SalePrice']
```

```python
df = df_org[columns]
```

```python
df.describe()
```

```python
X = df.drop(columns='SalePrice')
y = df['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
y_preds = lr.predict(X_test)
```

```python
root_mean_squared_error(y_test, y_preds)
```
